```python
# ==========================================
# 修复 ComfyUI 启动问题（不重新安装，不重下模型）
# 适用于 ComfyUI 已安装在 Google Drive 的情况
# ==========================================

import time
import os

# 工作路径（你的 ComfyUI 在 Drive）
COMFYUI_PATH = "/content/drive/MyDrive/ComfyUI"

print("🔄 正在清理残留进程...")

# 1. 杀掉所有相关进程
os.system("pkill -9 -f main.py 2>/dev/null")
os.system("pkill -9 -f cloudflared 2>/dev/null")
os.system("pkill -9 -f lt 2>/dev/null")
time.sleep(2)

# 2. 删除数据库锁文件（避免 "database is locked"）
os.chdir(COMFYUI_PATH)
os.system("rm -f user/comfyui.db-lock 2>/dev/null")
os.system("rm -f user/comfyui.db-shm 2>/dev/null")
os.system("rm -f user/comfyui.db-wal 2>/dev/null")

print("✅ 进程已清理，锁文件已删除")

# 3. 重新启动 ComfyUI（绑定 0.0.0.0，前台日志可见）
print("🚀 正在启动 ComfyUI...")
os.chdir(COMFYUI_PATH)
# 用 nohup 后台运行，但将输出重定向到日志文件，方便查看
os.system("nohup python main.py --listen 0.0.0.0 --port 8188 > comfyui.log 2>&1 &")
time.sleep(8)

# 4. 检查端口是否真正监听
print("🔍 检查端口状态...")
port_check = os.popen("netstat -tlnp | grep 8188").read()
if "0.0.0.0:8188" in port_check:
    print("✅ ComfyUI 已成功监听 8188 端口")
else:
    print("❌ ComfyUI 未监听 8188，请查看日志：!cat comfyui.log")
    print("日志内容：")
    os.system("tail -20 comfyui.log")

# 5. 启动 localtunnel（如果尚未安装则安装）
print("🔗 正在安装/启动 localtunnel...")
os.system("npm install -g localtunnel 2>/dev/null")

# 6. 前台运行 localtunnel（会输出链接）
print("⏳ 等待链接生成...")
os.system("lt --port 8188")
```

## CF Tunnel

```python
import subprocess
import threading
import time
import socket
import urllib.request

def iframe_thread(port):
  while True:
      time.sleep(0.5)
      sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
      result = sock.connect_ex(('127.0.0.1', port))
      if result == 0:
        break
      sock.close()
  print("\nComfyUI finished loading, trying to launch cloudflared (if it gets stuck here cloudflared is having issues)\n")

  p = subprocess.Popen(["cloudflared", "tunnel", "--url", "http://127.0.0.1:{}".format(port)], stdout=subprocess.PIPE, stderr=subprocess.PIPE)
  for line in p.stderr:
    l = line.decode()
    if "trycloudflare.com " in l:
      print("This is the URL to access ComfyUI:", l[l.find("http"):], end='')
    #print(l, end='')


threading.Thread(target=iframe_thread, daemon=True, args=(8188,)).start()

!python main.py --dont-print-server
```
